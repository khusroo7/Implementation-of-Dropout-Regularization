```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings 
warnings.filterwarnings('ignore')

df=pd.read_csv("./sonar_dataset.csv" , header=None)
df.sample(5)
```

```python
df.shape
```

```python
df.isna().sum() # checking if there are any null values present or not
```

```python
df[60].value_counts()
```

```python
# doing one hot encoding to change the target variable into 0 and 1
df[60].replace({'R':1, 'M':0}, inplace=True) # R--> 1 AND M--> 0
```

```python
df.sample(5)
```

```python
X=df.drop(60, axis='columns')
y=df[60]
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.25, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# craeting the neural network without adding the dropout layer
model=keras.Sequential([
    keras.layers.Dense(60, activation='relu'),
    keras.layers.Dense(60, activation='relu'),
    keras.layers.Dense(60, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
    
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred=model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer i.e 0 and 1
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
# now creating the model by adding a dropout layer after every hidden layer by dropping 50% of random neurons in hidden layer
modeld=keras.Sequential([
    keras.layers.Dense(60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid'),
])

modeld.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(X_test,y_test) # we can see that the neural netowrk without the dropout layer had less accuracy in the test phase i.e 0.73 whereas after adding the dropout layer the accuracy is increased in the test phase i.e 0.78
```

```python
y_pred=modeld.predict(X_test).reshape(-1)
y_pred=np.round(y_pred)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))  
```

```python
# f1 score without dropout layer was 0.77 and 0.72 but after that with dropout layer the f1 score was 0.82 and 0.74
```
